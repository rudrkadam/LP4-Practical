```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from tensorflow.keras.layers import Conv2D,MaxPooling2D,BatchNormalization,Dense,Flatten,Dropout
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential
```

```python
train = pd.read_csv('train_data.csv')
```

```python
test = pd.read_csv('test_data.csv')
```

```python
train
```

```python
x_train = train.drop(['label'] , axis = 1).values
```

```python
x_train = x_train.reshape(50000,32,32,3).astype('float32')/255
```

```python
x_train.shape
```

```python
y_train = train['label'].values
```

```python
y_train.shape
```

```python
test
```

```python
x_test = test.drop(['label'] , axis = 1).values
```

```python
x_test = x_test.reshape(10000,32,32,3).astype('float32')/255
```

```python
x_test.shape
```

```python
y_test = test['label'].values
```

```python
y_test.shape
```

```python
model = Sequential([
    Conv2D((64),(3,3),input_shape = (32,32,3),activation = 'relu'),
    BatchNormalization(),
    Conv2D((32),(3,3),input_shape = (32,32,3),activation = 'relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64,activation = 'relu'),
    Dropout(0.2),
    Dense(10,activation = 'softmax')
])
```

```python
model.compile(optimizer = SGD(learning_rate = 0.01) , loss = "sparse_categorical_crossentropy" , metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,batch_size = 64,epochs = 11,validation_split = 0.2)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
```

```python
loss
```

```python
accuracy
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.argmax(y_pred,1)
```

```python
y_pred
```

```python
csr = classification_report(y_test,y_pred)
```

```python
print(csr)
```
