```python
file_path = 'CBOW.txt'

with open(file_path,'r') as file :
    file_contents = file.read()
```

```python
file_contents
```

```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = file_contents.split('.')
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
total_words = len(tokenizer.word_index) + 1
tokenized_sentences = tokenizer.texts_to_sequences(sentences)
```

```python
tokenized_sentences
```

```python
window_size = 3

data = pad_sequences([[sentence[j] for j in range(i - window_size,i+window_size+1) if i!=j and j >=0 and j < len(sentence)]
                      for sentence in tokenized_sentences for i in range(len(sentence))
                     ])

labels = np.array([target_word for sentence in tokenized_sentences for target_word in sentence])
```

```python
model = models.Sequential([
    layers.Embedding(input_dim = total_words,output_dim = 50,input_length = 6),
    layers.GlobalAveragePooling1D(),
    layers.Dense(total_words,activation = 'softmax')
])
```

```python
model.compile(optimizer = 'adam' , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])
```

```python
history = model.fit(data,labels,epochs = 200,verbose = 1)
```

```python
model.summary()
```

```python
word_embeddings = model.get_weights()[0]
```

```python
target_word = 'virus'
target_embeddings = word_embeddings[tokenizer.word_index[target_word]]
```

```python
similarities = cosine_similarity(target_embeddings.reshape(1,-1),word_embeddings)[0]
```

```python
top_similar_words_indices = similarities.argsort()[::-1][:5]
```

```python
similar_words = [word for word,idx in tokenizer.word_index.items() if idx in top_similar_words_indices]
```
